# 7-2

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# 7-3

```python
df['Description'] = df['Description'].str.strip()
df
```

# 7-4

```python
df.dropna(subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df
```

# 7-5

```python
df = df[~df['InvoiceNo'].str.contains('C')]
df
```

# 7-6

```python
basket=(df[df['Country']=="France"].groupby(["InvoiceNo",'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket
```

# 7-7

```python
def Basket_0_1(x):
    if x > 0:
        return 1
    else:
        return 0

basket_set = basket.applymap(Basket_0_1)
```

# 7-8

```python
basket_set.drop('POSTAGE', inplace=True, axis=1)
basket_set
```

# 7-9

```python
frequent_itemsets = apriori(basket_set, min_support=0.07, use_colnames=True)
frequent_itemsets
```

# 7-10

```python
rules = association_rules(frequent_itemsets, metric="lift")
rules
```

# 7-11

```python
rules[ (rules['lift'] >= 6) & (rules['confidence'] >= 0.8) ]
```

# 2-4

```python
X = B.drop("Class variable (0 or 1)", axis=1)
y = B["Class variable (0 or 1)"]
```

# 2-5

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

# 2-6

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import tree
classifier = DecisionTreeClassifier(criterion='entropy',max_features=4,max_depth=5)
classifier.fit(X_train, y_train)
```

# 2-7

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('Testing-set accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
print('Training set score: {:.4f}'.format(classifier.score(X_train, y_train)))
print('Test set score: {:.4f}'.format(classifier.score(X_test, y_test)))
```

# 2-8

```python
a=[]
b=[]
for i in range(3,10):
    print(i)
    classifier = DecisionTreeClassifier(criterion='entropy',max_depth=i)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    #print(classification_report(y_test, y_pred))
    a.append(classifier.score(X_train, y_train))
    b.append(classifier.score(X_test, y_test))
    print('Training set score: {:.4f}'.format(classifier.score(X_train, y_train)))
    print('Test set score: {:.4f}'.format(classifier.score(X_test, y_test)))
plt.figure(figsize=(12, 6))
plt.plot(range(3, 10), a,linestyle='dashed', marker='o',markerfacecolor='blue', markersize=10,label="train")
plt.plot(range(3, 10), b, linestyle='dashed', marker='o',markerfacecolor='green', markersize=10,label="test")
plt.title('Accuracy')
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
plt.legend()
```

```python
X = B[["Plasma glucose concentration a 2 hours in an oral glucose tolerance test","Body mass index (weight in kg/(height in m)^2)","Diabetes pedigree function","Age (years)"]]
y = B["Class variable (0 or 1)"]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=2)
from sklearn import tree
classifier = DecisionTreeClassifier(criterion='entropy',max_features=3,max_depth=5)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('Testing-set accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
print('Training set score: {:.4f}'.format(classifier.score(X_train, y_train)))
print('Test set score: {:.4f}'.format(classifier.score(X_test, y_test)))
```

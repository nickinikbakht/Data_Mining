```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import pandas as pd
boston = load_boston()
```

```python
print("keys: ",boston.keys(),"\n")
print("feature_names: ",boston.feature_names)
```

# 5-1

```python
print(boston.DESCR)
```

# 5-2

```python
df = pd.DataFrame(boston.data, columns = boston.feature_names)
df.head()
```

# 5-3

```python
df['Price']=boston.target
df.head()
```

```python
b = df.values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(df)
standardized = scaler.transform(df)
df= pd.DataFrame(standardized,columns=["CRIM",'ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT',"Price"])
df
```

```python
df.to_csv('Boston.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20,20))
sns.set(font_scale=1)
sns.heatmap(df.corr(), center=0, cmap='Blues',annot=True)
```

```python
import seaborn as sns
sns.set(style="ticks", color_codes=True)
sns.pairplot(df)
```

# 5-4

```python
X=df[boston.feature_names]
y=df['Price']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# 5-5

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
pred_y = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print('Coefficient of determination: %.2f'% r2_score(y_test, pred_y))
```

# 5-6

```python
from sklearn import metrics
print("MSE: ",metrics.mean_squared_error(y_test, pred_y))
```

```python
metrics.mean_absolute_error(y_test, pred_y)
```

```python
reg.score(X_test, y_test)
```

# 5-7

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(reg, boston.data, boston.target, cv=5)
print("scores: ",scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

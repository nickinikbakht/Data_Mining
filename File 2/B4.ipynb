```python
from IPython.display import Image 
import pandas as pd
from sklearn import datasets
from scipy.cluster.hierarchy import dendrogram, linkage ,fcluster
from scipy.spatial.distance import pdist
from matplotlib import pyplot as plt
iris = datasets.load_iris()
print(iris.DESCR)
df= pd.DataFrame(iris.data,columns=iris.feature_names)
```

# 4-1

```python
distanceMatrix = pdist(iris.data,)
Z=linkage(distanceMatrix,'complete')
Z
```

# 4-2

```python
plt.figure(figsize=(25, 15),edgecolor='blue')
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('iris data')
plt.ylabel('distance')
dendrogram(Z)
plt.show()
```

# 4-3

```python
plt.figure(figsize=(20, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('iris data')
plt.ylabel('distance')
dendrogram(Z)
plt.axhline(y=6, color='black')
plt.show()
fcluster(Z, t=6, criterion='distance')
```

# 4-4

```python
plt.figure(figsize=(20, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('iris data')
plt.ylabel('distance')
dendrogram(Z)
plt.axhline(y=3, color='black')
plt.show()
fcluster(Z, t=3, criterion='distance')
```

```python
import seaborn as sns

df1=df
df1['Species']=iris.target
df1 = df.reset_index()
df1['id'] = df1.index

f, axes = plt.subplots(1,4,figsize=(15,5))
y =iris.feature_names
ss=fcluster(Z, t=3, criterion='distance')
ax = 0
for y in y:
    sns.scatterplot(x = 'id', y =y, hue = ss, data = df1, ax = axes[ax])
    ax = ax + 1
```
